## Stemming and Lemmatization

```python
from nltk import word_tokenize
import nltk
import re
import string
punch=set(string.punctuation)
```

```python
def clean(text):
    text=text.lower()
    text=re.sub(r'[^a-zA-Z]'," ",text)
    text="".join([i for i in text if i not in punch])
    stopfree = " ".join([i for i in text.split() if (len(i)>1 and i not in stopwords)])
    return stopfree.split()
```

# TfidfVectorizer

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(max_df=1.0,
                             min_df=0.0,
                             tokenizer=clean,
                             lowercase=False)
Tfidf_matrix = vectorizer.fit_transform(x)
```

# CountVectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer
Vectorizer = CountVectorizer(max_df=1.0,
                             min_df=0.0,
                             tokenizer=clean,
                             lowercase=False)
Tfidf_matrix_1 = Vectorizer.fit_transform(x)
```

# HashingVectorizer

```python
from sklearn.feature_extraction.text import HashingVectorizer
Vectorizer = HashingVectorizer(ngram_range=(0,1),
                               tokenizer=clean,
                               lowercase=False)
Tfidf_matrix_2 = Vectorizer.fit_transform(x)
```

# KMeans Clustering

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters=4,
               max_iter=300,
               tol=0.001,
               algorithm='auto')
model.fit(Tfidf_matrix)
labels = model.predict(Tfidf_matrix)
model.cluster_centers_
#print (model)
#print (Tfidf_matrix)
#print (x)
```

# Silhouette Score

```python
import sklearn
sklearn.metrics.silhouette_score(Tfidf_matrix, labels, metric='euclidean', sample_size=None, random_state=None)
```
